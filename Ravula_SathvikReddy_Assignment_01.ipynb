# Question 1 (10 points).

The formula that describes the volume V of a hemisphere with radius r is the following:

$ V=\frac{2}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of even numbers from 1-20. Output the result in the following format:

	r	V
	2	…
	4	…
	6	…
	.
	.
	.
	20	…


```python
# Your code here

import math

print("r\tV")  # Header
for r in range(2, 21, 2):  # Looping through even numbers from 2 to 20
    V = (2/3) * math.pi * (r ** 3)  # Volume of hemisphere formula
    print(f"{r}\t{V:.2f}")  # Printing formatted output
```

(b) Arrange the generated numbers in ascending order. Display the sorted list.

(c) Extend the program to arrange the numbers in descending order as well. Present the sorted list accordingly.

(d) Compute and output the average of the entire group of generated numbers.

(e) Enhance the program to calculate and display the average of the first 6 numbers from both the ascending and descending sorted lists.

```python

import random

def generate_random_numbers(count, start, end):
    return [random.randint(start, end) for _ in range(count)]

def compute_average(numbers):
    return sum(numbers) / len(numbers)

# (a) Generate 12 random numbers within the range of 50 to 150
random_numbers = generate_random_numbers(12, 50, 150)
print("Generated Numbers:", random_numbers)

# (b) Sort in ascending order
ascending_numbers = sorted(random_numbers)
print("Ascending Order:", ascending_numbers)

# (c) Sort in descending order
descending_numbers = sorted(random_numbers, reverse=True)
print("Descending Order:", descending_numbers)

# (d) Compute the average of all numbers
overall_average = compute_average(random_numbers)
print("Overall Average:", overall_average)

# (e) Compute the average of the first 6 numbers from both sorted lists
ascending_avg = compute_average(ascending_numbers[:6])
descending_avg = compute_average(descending_numbers[:6])
print("Average of first 6 numbers in Ascending Order:", ascending_avg)
print("Average of first 6 numbers in Descending Order:", descending_avg)
```
