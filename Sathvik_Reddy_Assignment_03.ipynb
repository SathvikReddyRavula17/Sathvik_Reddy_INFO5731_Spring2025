```python
import pandas as pd
import numpy as np

# Load the dataset
dataset_csv = pd.read_csv('/content/sample_data/cleaned_text_data (2).csv')
dataset_cleaned = dataset_csv.dropna(subset=['clean_abstract'])

# Calculate term frequency (TF)
tf_result = (dataset_cleaned['clean_abstract']).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis=0).reset_index()
tf_result.columns = ['words', 'tf']

# Calculate inverse document frequency (IDF)
N = dataset_csv.shape[0]  # Total number of documents
for i, word in enumerate(tf_result['words']):
    df = len(dataset_cleaned[dataset_cleaned['clean_abstract'].str.contains(word)])  # Number of documents containing the word
    idf = np.log(N / df)
    tf_result.loc[i, 'idf'] = idf

# Calculate term frequency-inverse document frequency (TF-IDF)
tf_result['tf_idf'] = tf_result['tf'] * tf_result['idf']

# Display the results
tf_result.head()
```

```python
import pandas as pd
from gensim.models import Word2Vec
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Load the CSV file
df = pd.read_csv("/content/sample_data/cleaned_text_data (2).csv")

# Convert the 'clean_abstract' column to string and drop any NaN values
df['clean_abstract'] = df['clean_abstract'].astype(str).dropna()

# Train a Word2Vec model with sentences from the 'clean_abstract' column
model = Word2Vec(sentences=df['clean_abstract'].str.split(), vector_size=300, window=5, min_count=5, workers=4)

# Perform PCA to reduce dimensionality of word embeddings to 2D
pca = PCA(n_components=2)
embeddings_2d = pca.fit_transform(model.wv.vectors)

# Plot word embeddings in 2D
plt.figure(figsize=(10, 10))
plt.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1], marker='o', alpha=0.5)
for i, word in enumerate(model.wv.index_to_key):
    plt.annotate(word, xy=(embeddings_2d[i, 0], embeddings_2d[i, 1]), fontsize=8)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Word Embeddings Visualization from Cleaned Abstracts')
plt.show()
```
